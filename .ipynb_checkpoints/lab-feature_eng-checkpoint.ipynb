```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
#shape
spaceship.shape 
```

**Check for data types**

```python
#types
spaceship.dtypes
```

**Check for missing values**

```python
#null values
spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#drop rows containing any missing value (be careful, create a new df otherwise this does not work)
spaceship_clean = spaceship.dropna(axis="rows")
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here
spaceship_clean['Cabin']=spaceship_clean['Cabin'].str[0]
```

- Drop PassengerId and Name

```python
#Drop PassengerId and Name
spaceship_clean = spaceship_clean.drop(columns = ["PassengerId","Name"])
```

```python
#non numerical columns
X = spaceship_clean.select_dtypes(include="object") 
y = spaceship_clean["Transported"] 
```

**Perform Train Test Split**

```python
#Perform Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0)
```

- For non-numerical columns, do dummies.

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop='first') # To avoid having an sparse_matrix as output

# X train dummies
ohe.fit(X_train) # The .fit() method determines the unique values of each column
X_train_trans_np = ohe.transform(X_train)
X_train_trans_np
```

```python
# X train as df
X_train_trans_df = pd.DataFrame(X_train_trans_np, columns=ohe.get_feature_names_out(), index=X_train.index)
X_train_trans_df
```

```python
# X test dummies
X_test_trans_np = ohe.transform(X_test)
# X_test_trans_np
```

```python
# X test as dataframe
X_test_trans_df = pd.DataFrame(X_test_trans_np, columns=ohe.get_feature_names_out(), index=X_test.index)
X_test_trans_df
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#KNN Classificator
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5) # 5 by default neghbours according to documentation

# train the model
knn.fit(X_train_trans_df, y_train)
```

- Evaluate your model's performance. Comment it

```python
print(f"The accuracy the model is {knn.score(X_test_trans_df, y_test)*100: .2f}%")
```

The accuracy of the model is 65%. This result is unsufficient ton consider the model as effective.
